```python
#Import libarary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Count plot

```python
#Impor the data
df=sns.load_dataset("titanic")
```

```python
sns.countplot(data=df,x="class",hue='class',order=["First","Second","Third"])
plt.title("Class count")
plt.show()
```

```python
for i in df.columns:
    sns.countplot(data=df,x=i)
    plt.show()
```

KDE plot 
how the data is distributed

```python
for i in df.select_dtypes(include="number").columns:
    sns.kdeplot(data=df,x=i)
    plt.show()
```

Histogram

```python
sns.histplot(data=df,x="age")
```

Boxplot
for checking the outlier

```python
sns.boxplot(data=df,x="age")
```

scatter plot
Relation ship between the two variable

```python
sns.scatterplot(data=df,x="age",y="fare",hue="class")
```

Line plot


```python
sns.lineplot(data=df,x="age",y="fare")
plt.show()
```

```python
sns.lineplot(data=df,x="age",y="fare",hue="survived")
plt.show()
```

Heat map


```python
sns.heatmap(df.select_dtypes(include="number").corr(),annot=True)
```

```python
sns.barplot(data=df,x="survived",y="age")
plt.show()
```

Pie chart

```python
df=sns.load_dataset("iris")
df=df.sample(3,random_state=42)
df.head()
```

```python

plt.pie(x=df["sepal_length"],labels=df["species"],autopct="%1.1f%%",radius=1,rotatelabels=True)
plt.show()
```

```python
sns.barplot(data=df,x="species",y="sepal_length")
plt.show()
```
